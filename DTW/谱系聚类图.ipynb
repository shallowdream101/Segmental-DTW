```r
data=read.csv("corr.csv")
data=data[,-1]
h=hclust(as.dist(data) )
plot(h)
```

```r
colnames(data)=0:1480
```

```r
write.csv(colnames(data)[cutree(h,k=2) ==1],"第一类.csv")
```

```r
write.csv(colnames(data)[cutree(h,k=2) ==2],"第二类.csv")
```
